```python
import pandas as pd
import os
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
cur_dir = os.path.dirname(os.path.realpath('__file__'))
```

```python
df = pd.read_csv(cur_dir + '/data/cross_raw_data/' + 'bc_full_crossing_data.csv')
```

```python
df['Total Crossers'] = df['Train Passengers'] + df['Bus Passengers'] + df['Personal Vehicle Passengers'] + df['Pedestrians']
```

```python
df_2013 = df[df['Year'] == 2013]
```

```python
year_total = df_2013.groupby(df['Year']).sum()
year_total
```

```python
# Create a list of colors (from iWantHue)
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552"]

For city in df_2013['City']:
    # Create a pie chart
    plt.pie(
        # using data total)arrests
        df_2013['Total Crossers'],
        # with the labels being officer names
        labels=df_2013['City'],
        # with no shadows
        shadow=False,
        # with colors
        colors=colors,
        # with the start angle at 90%
        startangle=90,
        # with the percent listed as a fraction
        autopct='%1.1f%%',
        )

    # View the plot
    plt.tight_layout()
    plt.show()
```
